```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
```

```python
# Load the datasets
customers_df = pd.read_csv(r"C:\Users\Hp\Downloads\Customers.csv")  # Customer demographic data
transactions_df = pd.read_csv(r"C:\Users\Hp\Downloads\Transactions.csv")  # Transaction history
```

```python
# Preprocessing the customer data
# Ensure that 'CustomerID' is present in the Customers dataframe
if 'CustomerID' not in customers_df.columns:
    raise ValueError("CustomerID column not found in the customer data.")
```

```python
# Ensure 'CustomerID' exists in transactions_df and handle it
if 'CustomerID' not in transactions_df.columns:
    raise ValueError("CustomerID column not found in the transaction data.")
```

```python

# Creating a pivot table for transaction history (CustomerID vs ProductID)
transaction_pivot = transactions_df.pivot_table(index='CustomerID', columns='ProductID', aggfunc='sum', fill_value=0)
transaction_pivot
```

```python
# Create a dictionary to store the top 3 similar customers for each customer
lookalike_dict = {}
```

```python
# Prepare the output for Lookalike.csv
lookalike_data = []
for cust_id, similar_customers in lookalike_dict.items():
    for similar_cust_id, score in similar_customers:
        lookalike_data.append([cust_id, similar_cust_id, score])
```

```python
lookalike_df = pd.DataFrame(lookalike_data, columns=['CustomerID', 'LookalikeCustomerID', 'SimilarityScore'])
```
